# Treating Missing Values In Python

```python
#missing value is represented by NAN. 
#Using numpy.nan to create missing values in cars DataFrame
cars.loc[1,'hp'] = '?'
cars.loc[3,'mpg'] = np.nan
cars.loc[2,'mpg'] = np.nan
cars.loc[4,'hp'] = np.nan
cars.loc[4,'wt'] = np.nan
cars.head()
```

```python
cars.info()
```

```python
# \ is called as escape char, which means to consider char following
# as char not special one.
cars.hp.str.contains('/?').sum()
```

```python
cars.hp.replace('?',np.nan,inplace=True)
```

```python
cars.head()
```

```python
cars.hp.str.contains('np.NaN').sum()
```

```python
cars.head()
```

```python
cars.loc[1,'wt'] = np.nan
cars.head()
```

```python
cars.isnull()
```

```python
cars.dropna() #drop the rows with null values
```

## Filling missing value with fillna() function

```python
cars.isnull().sum()
```

```python
cars.head()
```

```python
# filling missing value with constant. 
# This is not recommended normally as it doesn't approximate nearest value of missing value,resulting poor analysis
cars.fillna(cars.mean())
```

```python
#Treating missing value with forward filling.
#filling with value above the misssing value
cars.fillna(method='bfill')  #the value is taken fron bottom
```

```python
cars.head()
```

```python
cars.loc[2,:] = np.nan
```
